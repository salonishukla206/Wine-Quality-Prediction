## Wine quality prediction

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'winequality-red.csv')
df
```

```python
df.describe()
```

```python
#Creating a correaltion heatmap using seabborn
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr())
plt.show()
```

```python
#Checking out for null values 
print(df.isnull().sum())
```

```python
#Storing the feature to be predicted, ie, wine quality
y= df.iloc[:,-1]
y
```

```python
#Creating the predictor featur datafrme
x = df.iloc[:,0:11]
x
```

```python
from sklearn.model_selection import train_test_split
#Splitting the data into train and test in 7:3 ratio
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression

#Using the simple linear regression model for fitting the data
z = LinearRegression()
z.fit(x_train,y_train)
```

```python
#coeffiecint values of the selected features 
z.coef_
```

```python
test_pred = z.predict(x_test)
test_pred
```

```python
# rounding off the predicted values for test set
#Since the categories are integers
predicted_data = np.round_(test_pred)
predicted_data
```

```python
from sklearn import metrics
#Printing out the various metries for the predicted data
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, test_pred)))
```
